```python
#creamos la tabla personas
#cursor.execute("CREATE DOMAIN banco.""TIPO_DOC"" AS character varying(10) CHECK (VALUE IN ('NIF','NIE','PASAPORTE'));")
#cursor.execute("CREATE DOMAIN banco.""TIPO_PERSONA"" AS character varying(10) CHECK (VALUE IN ('FISICA','JURIDICA'));")
cursor.execute("CREATE TABLE IF NOT EXISTS banco.""PERSONAS"" \
( \
    ""NUM_PERSONA"" character varying(9) NOT NULL, \
    ""ENTIDAD""  character varying(4) NOT NULL, \
    ""TIPO_DOCUMENTO"" banco.""TIPO_DOC"" NOT NULL, \
    ""DOCUMENTO"" character varying(10) NOT NULL, \
    ""TIPO_PERSONA"" banco.""TIPO_PERSONA"" NOT NULL, \
    ""NOMBRE"" character varying(50) NOT NULL, \
    ""APELLIDO1"" character varying(50) NOT NULL, \
    ""APELLIDO2"" character varying(50), \
    ""FC_NACIMIENTO"" date NOT NULL, \
    ""PAIS_NACIMIENTO"" character varying(3) NOT NULL, \
    ""FC_ALTA_CLIENTE"" date NOT NULL, \
    ""FC_BAJA_CLIENT"" date, \
    CONSTRAINT PK_PERSONAS PRIMARY KEY (""NUM_PERSONA"", ""ENTIDAD"") \
);")
conexion.commit()
```

```python
# Creamos la tabla contratos
#cursor.execute("CREATE DOMAIN banco.""TIPO_CONT"" AS character varying(10) CHECK (VALUE IN ('CUENTAS','PASIVO','PRESTAMOS','TARJETAS'));")
#cursor.execute("CREATE DOMAIN banco.""TIPO_REL"" AS character varying(12) CHECK (VALUE IN ('TITULAR','AUTORIZADO','BENEFICIARIO','APODERADO'));")
cursor.execute("CREATE TABLE IF NOT EXISTS banco.""CONTRATOS"" \
( \
    ""ENTIDAD_PERSONA"" character varying(4) NOT NULL, \
    ""NUM_PERSONA"" character varying(9) NOT NULL, \
    ""ENTIDAD"" character varying(4) NOT NULL, \
    ""CENTRO"" character varying(4) NOT NULL, \
    ""CUENTA"" character varying(10) NOT NULL, \
    ""TIPO_CONTRATO"" banco.""TIPO_CONT"" NOT NULL, \
    ""FC_ALTA_CONTRATO"" date, \
    ""FC_BAJA_CONTRATO"" date, \
    ""FC_ALTA_RELACION"" date, \
    ""FC_BAJA_RELACION"" date, \
    ""TIPO_RELACION"" banco.""TIPO_REL"", \
    CONSTRAINT ""PK_CONTRATOS"" PRIMARY KEY (""ENTIDAD"",""CENTRO"",""CUENTA""), \
    CONSTRAINT ""FK_PERSONAS"" FOREIGN KEY (""NUM_PERSONA"", ""ENTIDAD_PERSONA"") \
            REFERENCES banco.PERSONAS (""NUM_PERSONA,ENTIDAD"") \
            ON DELETE CASCADE \
            ON UPDATE CASCADE \
);")
conexion.commit()
```

```python
# Creamos la tabla cuentas
#cursor.execute("CREATE DOMAIN banco.""TIPO"" AS character varying(10) CHECK (VALUE IN ('CORRIENTE','AHORRO','LIBRETA','CREDITO'));")
cursor.execute("CREATE TABLE IF NOT EXISTS banco.""CUENTAS"" \
( \
    ""ENTIDAD"" character varying(4) NOT NULL, \
    ""CENTRO"" character varying(4) NOT NULL, \
    ""CUENTA"" character varying(10) NOT NULL, \
    ""TIPO_CUENTA"" banco.""TIPO"" NOT NULL, \
    ""FC_ALTA_CONTRATO"" date NOT NULL, \
    ""FC_BAJA_CONTRATO"" date, \
    ""MONEDA"" character varying(3) NOT NULL, \
    ""NUM_INTERVINIENTES"" numeric(5), \
    CONSTRAINT ""PK_CUENTAS"" PRIMARY KEY (""ENTIDAD"",""CENTRO"", ""CUENTA""), \
    CONSTRAINT ""FK_CONTRATOS"" FOREIGN KEY (""ENTIDAD"", ""CENTRO"",""CUENTA"") \
        REFERENCES banco.""CONTRATOS"" (""ENTIDAD"", ""CENTRO"",""CUENTA"") \
        ON DELETE CASCADE \
        ON UPDATE CASCADE \
);")
conexion.commit()
```
